```python
#Importar as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from mlxtend.frequent_patterns import apriori, association_rules
warnings.filterwarnings("ignore")
```

```python
#Carregar o dataset
df = pd.read_csv("basket_supermercado_1000.csv")
```

```python
print(df)
```

```python
pd.set_option('display.max_columns', None)
df = df.drop(columns=['transacao'])
print(df.head())
```

```python
#Aplicar o algoritmo apriori
frequent_itemssets = apriori(
    df,
    min_support=0.2,
    use_colnames=True
)

print("itemsets/subconjunto frequentes: ")
print(frequent_itemssets.sort_values(by="support", ascending=True))
```

```python
#Gerar as regras de associação
rules = association_rules(
    frequent_itemssets,
    metric="confidence",
    min_threshold=0.5
)

print("Regras de associação: ")
pd.set_option("display.max_columns",None)
pd.set_option("display.width",None)
pd.set_option("display.max_colwidth",None)
print(rules)
```

```python
#Filtrar as regras relevantes
rules_filtered = rules[
    (rules['lift'] > 1) &
    (rules['confidence'] >= 0.6)
].sort_values(by='lift', ascending=False)

print("Regras relevantes")
print(rules_filtered[["antecedents","consequents","support","confidence","lift"]])
```

```python
#insights para o gerente
print("Insights para o gerente")

for _, rule in rules_filtered.iterrows():
  antecedentes = ", ".join(list(rule['antecedents']))
  consequentes = ", ".join(list(rule['consequents']))
  print(
      f"Cliente que compram: [{antecedentes}]"
      f"tendem a comprar: [{consequentes}]"
      f"(confiança=rule{rule['confidence']:.2f}, lift={rule['lift']:.2f})"
  )
```
